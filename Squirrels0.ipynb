```python
import pandas as pd

url = 'https://data.cityofnewyork.us/resource/vfnx-vebw.csv'

wildlife = pd.read_csv(url)
```

```python
wildlife.head()
```

```python
wildlife.info()
```

```python
(pd.to_numeric(wildlife['age'], errors='coerce') < 13)
```

```python
wildlife[wildlife['age'].isna()]
```

```python
wildlife_dropna = wildlife.dropna()
wildlife_dropna.head()
```

```python
wildlife_cleaned = wildlife.copy()
```

```python
wildlife_cleaned['age'].fillna('Adult', inplace=True)
```

```python
wildlife_cleaned['primary_fur_color'].fillna('Unknown', inplace=True)
```

```python
wildlife_cleaned['highlight_fur_color'].fillna('Unknown', inplace=True)
```

```python
wildlife_cleaned['other_interactions'].fillna('Unkown', inplace=True)
```
