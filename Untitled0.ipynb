```python
import matplotlib.pyplot as plt

wildlife['shift'] = wildlife['shift'].fillna('Unknown Shift')
wildlife['eating'] = wildlife['eating'].astype(str).str.lower().map({'true': True, 'false': False})
wildlife['eating'] = wildlife['eating'].fillna(False)

shift_caught_eating = wildlife.groupby('shift')['eating'].value_counts().unstack(fill_value=0)
shift_caught_eating.rename(columns={False: 'Not Eating', True: 'Eating'}, inplace=True)

plt.figure(figsize=(10,7))
shift_caught_eating.plot(kind='bar', ax=plt.gca())
plt.title('Shift Squirrels Got Caught Eating')
plt.xlabel('Shift')
plt.ylabel('Number of Squirrels')
plt.xticks(rotation=45, ha='right')
plt.legend(title='If They Are Eating')
plt.tight_layout()
plt.show()
```

```python
wildlife['foraging'] = wildlife['foraging'].astype(str).str.lower().map({'true': True, 'false': False})
wildlife['foraging'] = wildlife['foraging'].fillna(False)

shift_caught_eating = wildlife.groupby('shift')['foraging'].value_counts().unstack(fill_value=0)
shift_caught_eating.rename(columns={False: 'Not Foraging', True: 'Foraging'}, inplace=True)

plt.figure(figsize=(10,7))
shift_caught_eating.plot(kind='bar', ax=plt.gca())
plt.title('Shift Squirrels Were Foraging')
plt.xlabel('Shift')
plt.ylabel('Number of Squirrels')
plt.xticks(rotation=45, ha='right')
plt.legend(title='If They Are Foraging')
plt.tight_layout()
plt.show()
```

```python
#Calling Label Encoder
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest, chi2
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(wildlife['shift'])
#Chosing specific columns
selected_feature_cols = ['running', 'age', 'location', 'foraging']
X_wildlife_raw = wildlife[selected_feature_cols].copy()
#Dumies for X
X = pd.get_dummies(X_wildlife_raw, columns=selected_feature_cols, drop_first=True)
#chi feature selection
selector = SelectKBest(chi2, k=3)
X_new = selector.fit_transform(X,y)
selected_features = X.columns[selector.get_support(indices=True)]
discarded_features = X.columns[~selector.get_support(indices=True)]

print("Chi-Squared Test for Feature Selection:")
print('Selected Features:', selected_features)
print('Discarded Features:', discarded_features)
feature_scores = pd.DataFrame({'Feature': X.columns, 'Chi2 Score': selector.scores_, 'P-value': selector.pvalues_})
feature_scores = feature_scores.sort_values(by='Chi2 Score', ascending=False).reset_index(drop=True)
print(feature_scores)
```
